### This demonstrates some common probability distributions using the Numpy random library. These can be useful in Monte Carlo simulations.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Uniform distribution. 
uniform = plt.hist(np.random.uniform(10, 30, 1000), 
             bins=200,
             normed=True)
plt.show()
```

```python
# Normal or Gaussian distribution.  a.k.a "bell curve"
normal = plt.hist(np.random.normal(loc=0, scale=10, size=100000), 
             bins=200,
             normed=True)
plt.show()
```

```python
# Normal distribution again, with theoretical curve overlaid

mean = 0
std_dev = 1
n = np.random.normal(loc=mean, scale=std_dev, size=1000)

count, bins, ignored = plt.hist(n, 30, normed=True)
plt.plot(bins, 1/(std_dev * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mean)**2 / (2 * std_dev**2) ),
          linewidth=2, color='r')
plt.show()
```

```python
# Triangular distribution.  
tri = plt.hist(np.random.triangular(10, 14, 30, 1000), 
             bins=200,
             normed=True)
plt.show()
```

```python
# Beta distribution.  
# https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.random.beta.html#numpy.random.beta
# http://blog.josephmisiti.com/the-beta-distribution

beta = plt.hist(np.random.beta(1, 1, 100000), 
             bins=200,
             normed=True)
plt.show()
```
